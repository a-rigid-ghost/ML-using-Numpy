```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('train.csv')
X1 = data.iloc[:,0:1]
Y = data.iloc[:,1:2].values

data_pred = pd.read_csv('test.csv')
X_test1 = data_pred.iloc[:,0:1]
Y_test = data_pred.iloc[:,1:2].values
```

```python
plt.scatter(X1,Y)
plt.title("Training")
plt.xlabel("Feature")
plt.ylabel("Label")
plt.show()
```

```python
plt.scatter(X_test1,Y_test)
plt.title("Test")
plt.xlabel("Feature")
plt.ylabel("Label")
plt.show()
```
